```python
# Convertir las columnas de Elo a numérico
chess_data['WhiteElo'] = pd.to_numeric(chess_data['WhiteElo'], errors='coerce')
chess_data['BlackElo'] = pd.to_numeric(chess_data['BlackElo'], errors='coerce')

# Estadísticas descriptivas para las calificaciones ELO de los jugadores blancos y negros
elo_stats = chess_data[['WhiteElo', 'BlackElo']].describe()

elo_stats
```

```python
# Contar los resultados de las partidas
result_counts = chess_data['Result'].value_counts(normalize=True) * 100

# Renombrar los índices para una mejor interpretación
result_counts.index = ['Blancas ganan', 'Negras ganan', 'Empate']

result_counts
```

```python
# Las 10 aperturas más comunes
top_openings = chess_data['Opening'].value_counts().head(10)

top_openings
```

```python
# Agrupar los datos por Apertura y Resultado, y tomar las 10 aperturas más comunes
top_openings_results = (
    chess_data[chess_data['Opening'].isin(top_openings.index)]
    .groupby(['Opening', 'Result'])
    .size()
    .unstack(fill_value=0)
)

# Calcular las tasas de victoria, derrota y empate
top_openings_results['Total'] = top_openings_results.sum(axis=1)
top_openings_results['Blancas ganan (%)'] = top_openings_results['1-0'] / top_openings_results['Total'] * 100
top_openings_results['Negras ganan (%)'] = top_openings_results['0-1'] / top_openings_results['Total'] * 100
top_openings_results['Empate (%)'] = top_openings_results['1/2-1/2'] / top_openings_results['Total'] * 100

# Seleccionar solo las columnas de porcentaje
top_openings_results_percentage = top_openings_results[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

top_openings_results_percentage.sort_values('Blancas ganan (%)', ascending=False)
```

```python
# Definir los límites de los rangos de Elo
elo_bins = range(700, 2500, 200)

# Crear una columna con los rangos de Elo
chess_data['EloRange'] = pd.cut(chess_data['WhiteElo'], bins=elo_bins)

# Agrupar por rango de Elo y resultado, y calcular las tasas de victoria, derrota y empate
elo_range_results = (
    chess_data.groupby(['EloRange', 'Result'])
    .size()
    .unstack(fill_value=0)
)

elo_range_results['Total'] = elo_range_results.sum(axis=1)
elo_range_results['Blancas ganan (%)'] = elo_range_results['1-0'] / elo_range_results['Total'] * 100
elo_range_results['Negras ganan (%)'] = elo_range_results['0-1'] / elo_range_results['Total'] * 100
elo_range_results['Empate (%)'] = elo_range_results['1/2-1/2'] / elo_range_results['Total'] * 100

# Seleccionar solo las columnas de porcentaje
elo_range_results_percentage = elo_range_results[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

elo_range_results_percentage
```

```python
# Agrupar por control de tiempo y resultado
time_control_results = (
    chess_data.groupby(['TimeControl', 'Result'])
    .size()
    .unstack(fill_value=0)
)

# Calcular las tasas de victoria, derrota y empate
time_control_results['Total'] = time_control_results.sum(axis=1)
time_control_results['Blancas ganan (%)'] = time_control_results['1-0'] / time_control_results['Total'] * 100
time_control_results['Negras ganan (%)'] = time_control_results['0-1'] / time_control_results['Total'] * 100
time_control_results['Empate (%)'] = time_control_results['1/2-1/2'] / time_control_results['Total'] * 100

# Seleccionar solo las columnas de porcentaje y las 10 controles de tiempo más comunes
time_control_results_percentage = time_control_results.nlargest(10, 'Total')[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

time_control_results_percentage
```

```python
# Seleccionar las 5 aperturas más comunes
top_5_openings = chess_data['Opening'].value_counts().head(5).index

# Filtrar los datos para incluir solo las 5 aperturas más comunes
top_5_openings_data = chess_data[chess_data['Opening'].isin(top_5_openings)]

# Agrupar por apertura y resultado
top_5_openings_results = (
    top_5_openings_data.groupby(['Opening', 'Result'])
    .size()
    .unstack(fill_value=0)
)

# Calcular las tasas de victoria, derrota y empate
top_5_openings_results['Total'] = top_5_openings_results.sum(axis=1)
top_5_openings_results['Blancas ganan (%)'] = top_5_openings_results['1-0'] / top_5_openings_results['Total'] * 100
top_5_openings_results['Negras ganan (%)'] = top_5_openings_results['0-1'] / top_5_openings_results['Total'] * 100
top_5_openings_results['Empate (%)'] = top_5_openings_results['1/2-1/2'] / top_5_openings_results['Total'] * 100

# Seleccionar solo las columnas de porcentaje
top_5_openings_results_percentage = top_5_openings_results[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

top_5_openings_results_percentage
```

```python
# Seleccionar las 3 aperturas más comunes
top_3_openings = chess_data['Opening'].value_counts().head(3).index

# Filtrar los datos para incluir solo las 3 aperturas más comunes
top_3_openings_data = chess_data[chess_data['Opening'].isin(top_3_openings)]

# Agrupar por rango de Elo, apertura y resultado
top_3_openings_by_elo_results = (
    top_3_openings_data.groupby(['EloRange', 'Opening', 'Result'])
    .size()
    .unstack(fill_value=0)
)

# Calcular las tasas de victoria, derrota y empate
top_3_openings_by_elo_results['Total'] = top_3_openings_by_elo_results.sum(axis=1)
top_3_openings_by_elo_results['Blancas ganan (%)'] = top_3_openings_by_elo_results['1-0'] / top_3_openings_by_elo_results['Total'] * 100
top_3_openings_by_elo_results['Negras ganan (%)'] = top_3_openings_by_elo_results['0-1'] / top_3_openings_by_elo_results['Total'] * 100
top_3_openings_by_elo_results['Empate (%)'] = top_3_openings_by_elo_results['1/2-1/2'] / top_3_openings_by_elo_results['Total'] * 100

# Seleccionar solo las columnas de porcentaje
top_3_openings_by_elo_percentage = top_3_openings_by_elo_results[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

top_3_openings_by_elo_percentage.reset_index()
```

```python
# Agrupar por rango de Elo, control de tiempo y resultado
time_control_by_elo_results = chess_data.groupby(['EloRange', 'TimeControl', 'Result']).size().reset_index(name='Count')

# Pivotar la tabla para tener los resultados como columnas
time_control_by_elo_pivot = time_control_by_elo_results.pivot_table(index=['EloRange', 'TimeControl'], columns='Result', values='Count', fill_value=0)

# Calcular las tasas de victoria, derrota y empate
time_control_by_elo_pivot['Total'] = time_control_by_elo_pivot.sum(axis=1)
time_control_by_elo_pivot['Blancas ganan (%)'] = time_control_by_elo_pivot['1-0'] / time_control_by_elo_pivot['Total'] * 100
time_control_by_elo_pivot['Negras ganan (%)'] = time_control_by_elo_pivot['0-1'] / time_control_by_elo_pivot['Total'] * 100
time_control_by_elo_pivot['Empate (%)'] = time_control_by_elo_pivot['1/2-1/2'] / time_control_by_elo_pivot['Total'] * 100

# Seleccionar solo las columnas relevantes
time_control_by_elo_percentage = time_control_by_elo_pivot[['Blancas ganan (%)', 'Negras ganan (%)', 'Empate (%)']]

# Mostrar los primeros 10 registros
time_control_by_elo_percentage.head(10)
```
